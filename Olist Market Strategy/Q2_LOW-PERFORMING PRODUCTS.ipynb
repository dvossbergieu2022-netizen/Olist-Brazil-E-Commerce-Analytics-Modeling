```python
# regression with no categories + interaction effects + polynomials (squared)

data['interaction_photos_name_length'] = data['product_photos_qty'] * data['product_name_lenght']
data['interaction_photos_description_length'] = data['product_photos_qty'] * data['product_description_lenght']
data['interaction_name_description_length'] = data['product_name_lenght'] * data['product_description_lenght']

data['photos_qty_squared'] = data['product_photos_qty'] ** 2
data['name_length_squared'] = data['product_name_lenght'] ** 2
data['description_length_squared'] = data['product_description_lenght'] ** 2

data['photos_qty_cubed'] = data['product_photos_qty'] ** 3
data['name_length_cubed'] = data['product_name_lenght'] ** 3
data['description_length_cubed'] = data['product_description_lenght'] ** 3

data['photos_qty_root'] = data['product_photos_qty'] ** 0.5
data['name_length_root'] = data['product_name_lenght'] ** 0.5
data['description_length_root'] = data['product_description_lenght'] ** 0.5

data['description_length_div1000'] = data['product_description_lenght'] / 1000

data['photos_present'] = (data['product_photos_qty'] >= 1).astype(int)

data['more_than_one_picture'] = (data['product_photos_qty'] >= 2).astype(int)
```

```python
# Create dummy variables
dummies = pd.get_dummies(data['product_category_name'], prefix='category')

# Add these dummy variables to your original data
data = pd.concat([data, dummies], axis=1)
```

```python
# Calculate the quartile thresholds
q1 = data['product_description_lenght'].quantile(0.25)  # 25th percentile
q2 = data['product_description_lenght'].quantile(0.50)  # Median (50th percentile)
q3 = data['product_description_lenght'].quantile(0.75)  # 75th percentile

# Create the categorical variable
def categorize_description_length(length):
    if length <= q1:
        return 0  # Lowest 25%
    elif q1 < length <= q2:
        return 0.5  # Between 25% and 50%
    elif q2 < length <= q3:
        return 1  # Between 50% and 75%
    else:
        return 2  # Top 25%

# Apply the function to create the new column
data['description_length_category'] = data['product_description_lenght'].apply(categorize_description_length)
```

## cross validation


```python
# k-fold


from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import statsmodels.api as sm

# Initialize k-fold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)  # Use 5 folds, you can adjust n_splits

# Store metrics for each fold
mse_scores = []
r2_scores = []

for train_index, test_index in kf.split(X4):
    # Split the data into training and testing sets
    X_train, X_test = X4.iloc[train_index], X4.iloc[test_index]
    y_train, y_test = y2.iloc[train_index], y2.iloc[test_index]

    # Add a constant to the predictor variables for the intercept
    X_train = sm.add_constant(X_train)
    X_test = sm.add_constant(X_test)

    # Fit the model
    model = sm.OLS(y_train, X_train).fit()

    # Predict on the test set
    y_pred = model.predict(X_test)

    # Calculate metrics
    mse_scores.append(mean_squared_error(y_test, y_pred))
    r2_scores.append(r2_score(y_test, y_pred))

# Average the metrics across all folds
avg_mse = np.mean(mse_scores)
avg_r2 = np.mean(r2_scores)

print(f"Average MSE across folds: {avg_mse}")
print(f"Average R2 across folds: {avg_r2}")
```

```python
# LOOCV

from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import statsmodels.api as sm

# Initialize LOOCV (Leave-One-Out Cross-Validation)
kf = KFold(n_splits=len(X4), shuffle=False)  # Each sample is a test set

# Store metrics for each iteration
mse_scores = []
r2_scores = []

# Loop through each split
for train_index, test_index in kf.split(X4):
    # Split the data into training and testing sets
    X_train, X_test = X4.iloc[train_index], X4.iloc[test_index]
    y_train, y_test = y2.iloc[train_index], y2.iloc[test_index]

    # Add a constant to the predictor variables for the intercept
    X_train = sm.add_constant(X_train)
    X_test = sm.add_constant(X_test)

    # Fit the model on the training set
    model = sm.OLS(y_train, X_train).fit()

    # Predict on the test set
    y_pred = model.predict(X_test)

    # Calculate metrics
    mse_scores.append(mean_squared_error(y_test, y_pred))
    r2_scores.append(r2_score(y_test, y_pred))

# Average the metrics across all folds
avg_mse = np.mean(mse_scores)
avg_r2 = np.mean(r2_scores)

print(f"Average MSE across folds: {avg_mse}")
print(f"Average R2 across folds: {avg_r2}")
```
